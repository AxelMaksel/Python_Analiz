```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('kc-house-data.csv',encoding='cp1251')

plt.figure(figsize=(20, 6))
plt.hist(df['price'], bins = 40)
plt.title('Распределение цен')
plt.xlabel('Цена')
plt.ylabel('количество')
plt.ticklabel_format(style='plain')
plt.show()
```

Видим, что большинство квартир соимостью до миллиона, а 
больше всего ценой около 400-500 тысяч, а площадь большинства квартир 1500-2500 кв.футов

```python
plt.figure(figsize=(20, 6))
plt.hist(df['sqft_living'], bins = 40)
plt.title('Распределение жилой площади')
plt.xlabel('площадь')
plt.ylabel('количество');
```

```python
plt.figure(figsize=(20, 6))
plt.hist(df['yr_built'], bins = 30)
plt.title('Распределение годов постройки')
plt.xlabel('площадь')
plt.ylabel('количество');
```

самый поппулярный для продажи годы постройки 2005-2010
