```python
data = pd.crosstab(df.Company, df.Cpu_Company)
data.drop(columns='Samsung', inplace=True)
data.head(2)
```

```python
plt.figure(figsize=(10, 4))
n_tick = np.arange(len(data['AMD']))
offset = 0.2
w = 0.4
plt.bar(n_tick - offset, data['AMD'], width=w)
plt.bar(n_tick + offset, data['Intel'], width=w)
plt.xticks(n_tick,data.index, rotation=70)
plt.legend(['AMD', 'Intel']);
```

```python
data = df[df['Cpu_Company'] != 'Samsung']
data = pd.crosstab(data.Company, data.Cpu_Company,normalize = 'index')
data
```

```python
plt.figure(figsize=(8, 3))
plt.bar(data.index, data['Intel'])
plt.bar(data.index, data['AMD'], bottom=data['Intel'])
plt.title('Ноутбуки')
plt.legend(['Intel', 'AMD'])
plt.xticks(rotation=50);
```

Постройте матрицу корреляций для таблицы

```python
df = pd.read_csv('laptops_10.csv')
df=df.select_dtypes(include=('float64', 'int64'))
cor_mat = df.corr()
df.head(7)
```

```python
cor_mat = round(cor_mat, 2)
cor_mat
```

```python
cor_mat[np.abs(cor_mat)<0.3]=0
cor_mat
```

```python
sns.heatmap(cor_mat, cmap='coolwarm', annot=True, linewidths=4);
```
