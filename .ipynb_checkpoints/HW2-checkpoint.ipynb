## Условие 3: 3 задача
Ответьте на несколько вопросов
### 3.1 В каком диапазоне изменяются стоимости недвижимости?
### 3.2 Какую долю в среднем занимают жилая площадь от всей площади по всем домам?
### 3.3 Как много домов с разными этажами в данных?
### 3.4 Насколько хорошие состояния у домов в данных?
### 3.5 Найдите года, когда построили первый дом, когда построили последний дом в данных?

```python
min = df['price'].min()
max = df['price'].max()
print("1:", min, max)
living_mean = df['sqft_living'].mean()
lot_mean = df['sqft_lot'].mean()
print("2: {percent:2.1f}%".format(percent=living_mean/ lot_mean * 100))
```

```python
floors = df[df['floors'] != 1].shape[0]
print("3:", floors)
print("4: среднее состояние домов {condition:2.1f} ".format(condition = df['condition'].mean()))
print("5: старый дом {old} года, новый дом {new} года".format(old = df['yr_built'].min(), new = df['yr_built'].max()))
```

## Условие 4: 4 задача
Ответьте на несколько вопросов

### 4.1 Сколько в среднем стоят дома, у которых 2 спальни?
### 4.2 Какая в среднем общая площадь домов, у которых стоимость больше 600 000?
### 4.3 Как много домов коснулся ремонт?
### 4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?


```python
bathrooms = df[df['bathrooms'] == 2]
print("1: среднее Цена домов {bathroom:2.1f} ".format(bathroom = bathrooms['price'].mean()))
lot = df[df['price'] > 600000]
print("2: среднея площадь домов {lot:2.1f} ".format(lot = lot['sqft_lot'].mean()))
print("3: домов с ремонтом", df[df['yr_renovated'] != 0].shape[0])
```

```python
grade4 = df[df['grade'] < 4]
grade10 = df[df['grade'] > 10]
price4 = grade4['price'].mean()
price10 = grade10['price'].mean()
print(f"Плохие дома стоят в среднем {price4:,.0f}\nхорошие дома стоят в среднем {price10:,.0f}")
```

## Условие 5: 5 задача
Ответьте на несколько вопросов
### 5.1 Выберите дом клиенту
Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?


```python
df[(df['bathrooms'] >= 3) & (df['sqft_basement'] != 0) & (df['waterfront'] != 0)].shape[0]
```

### 5.2 Выберите дом клиенту
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

```python
tmp = df[(df['condition'] == 5) & ((df['view'] == 4) | (df['waterfront'] != 0)) & (df['yr_built'] >= 1980)]
print("min", tmp['price'].min())
print("max", tmp['price'].max())
```

### 5.3 Выберите дом клиенту
Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?

```python
tmp = df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] <= 150000)]
tmp['condition'].describe()
```

Средняя оценка "три", но есть и дома в хорошем состоянии, в отличном нет домов

- id - Уникальный ID для каждого дома
- date - Дата продажи дома
- price - Стоимость продажи дома
- bedrooms - Кол-во спален
- bathrooms - Кол-во ванных комнат (0.5 - туалет без душа)
- sqft_living - Кв. метры жилые
- sqft_lot - Кв. метры общие
- floors - Кол-во этажей
- waterfront - Есть набержная или нет
- view - Значение от 0 до 4 насколько хороший вид
- condition - Значение от 1 до 5 насколько хорошее состояние
- grade - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна
- sqft_above - Кв. метры дома, которые находятся выше земли
- sqft_basement - Кв. метры дома, которые находятся ниже земли
- yr_built - Год постройки дома
- yr_renovated - Год ремонта дома
- zipcode - Индекс
- lat - Широтма
- long - Долгота
- sqft_living15 - Кв. метры жилой площади у 15 соседей
- sqft_lot15 - Кв. метры общей площади у 15 соседей
